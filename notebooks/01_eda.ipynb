# Incident Triage Assistant — Quick EDA

Run this notebook to get a feel for the dataset before running the pipeline.
It loads the cleaned data and shows class distributions, text lengths, and KB stats.

Pre-requisite: `python -m src.preprocess` has been run at least once.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT))

import pandas as pd
import matplotlib.pyplot as plt
from src.config import load_config

cfg = load_config()
inc = pd.read_csv(cfg.paths.processed_dir / 'incidents_clean.csv', low_memory=False)
kb = pd.read_csv(cfg.paths.processed_dir / 'kb_clean.csv', low_memory=False)
print(f'Incidents: {len(inc):,}    KBAs: {len(kb):,}')
```

## 1. Assignment group distribution

```python
ag_counts = inc['assignment_group'].value_counts()
print(ag_counts)
ag_counts.plot.bar(figsize=(8, 3.5), color='#185fa5')
plt.title('Tickets per assignment group')
plt.ylabel('count')
plt.tight_layout()
plt.show()
```

## 2. Top business services

```python
bs_counts = inc['business_service'].value_counts().head(20)
bs_counts.plot.barh(figsize=(8, 7), color='#0f6e56')
plt.title('Top 20 business services')
plt.xlabel('count')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## 3. Text length distribution

```python
lens = inc['text'].fillna('').str.len()
print(lens.describe())
lens.clip(upper=2000).hist(bins=40, color='#534AB7')
plt.title('Incident text length (clipped at 2000 chars)')
plt.xlabel('characters')
plt.ylabel('count')
plt.show()
```
